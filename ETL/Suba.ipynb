CREAMOS LA COLUMNA IDPLATAFORMA EN CADA DATAFRAME Y POSTERIORMENTE UNIRLAS EN UNA SOLA TABLA.

```python
df_amazon["idPlataforma1"] =   np.arange(1,len(df_amazon)+1)
df_amazon.idPlataforma1 = df_amazon.idPlataforma1.astype('str')
df_amazon["idPlataforma2"] =   "AMZN_"
df_amazon["idPlataforma"] = df_amazon.idPlataforma2 + df_amazon.idPlataforma1
```

```python
df_amazon.drop(["idPlataforma1","idPlataforma2"], axis= 1,inplace = True)

df_amazon.columns
```

```python
df_amazon.head(2)
```

```python
df_netflix["idPlataforma1"] =   np.arange(1,len(df_netflix)+1)
df_netflix.idPlataforma1 = df_netflix.idPlataforma1.astype('str')
df_netflix["idPlataforma2"] =   "NETF_"
df_netflix["idPlataforma"] = df_netflix.idPlataforma2 + df_netflix.idPlataforma1

```

```python
df_netflix.drop(["idPlataforma1","idPlataforma2"], axis= 1,inplace = True)
```

```python
df_netflix.head(2)
```

```python
df_disney["idPlataforma1"] =   np.arange(1,len(df_disney)+1)
df_disney.idPlataforma1 = df_disney.idPlataforma1.astype('str')
df_disney["idPlataforma2"] =   "DSNY_"
df_disney["idPlataforma"] = df_disney.idPlataforma2 + df_disney.idPlataforma1
```

```python
df_disney.drop(["idPlataforma1","idPlataforma2"], axis= 1,inplace = True)

df_disney.columns

```

```python
df_hulu["idPlataforma1"] =   np.arange(1,len(df_hulu)+1)
df_hulu.idPlataforma1 = df_hulu.idPlataforma1.astype('str')
df_hulu["idPlataforma2"] =   "HULU_"
df_hulu["idPlataforma"] = df_hulu.idPlataforma2 + df_hulu.idPlataforma1
```

```python
df_hulu.drop(["idPlataforma1","idPlataforma2"], axis= 1,inplace = True)

df_hulu.columns

```
